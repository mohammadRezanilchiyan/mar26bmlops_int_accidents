```python
# -*- coding: utf-8 -*-
import pandas as pd
from pathlib import Path

# --- Paths ---
DATA_DIR = Path("/Users/mohammadrezanilchiyan/projects/mar26bmlops_int_accidents/data")
OUTPUT_DIR = DATA_DIR / "processed"
OUTPUT_DIR.mkdir(exist_ok=True)
OUTPUT_FILE = OUTPUT_DIR / "all_accidents_2005_2020.csv"

YEARS = range(2005, 2021)
all_data = pd.DataFrame()

# --- Helper function to find correct column ---
def find_column(df, possible_names):
    """Return the first column in df that matches any name in possible_names (case-insensitive)."""
    df_cols_lower = [c.strip().lower() for c in df.columns]
    for name in possible_names:
        if name.lower() in df_cols_lower:
            return df.columns[df_cols_lower.index(name.lower())]
    return None

for year in YEARS:
    print(f"\nProcessing year: {year}")
    
    # File paths
    usagers_file = DATA_DIR / "usagers" / f"usagers-{year}.csv"
    caract_file = DATA_DIR / "caracteristiques" / f"caracteristiques-{year}.csv"
    veh_file = DATA_DIR / "vehicules" / f"vehicules-{year}.csv"
    lieux_file = DATA_DIR / "lieux" / f"lieux-{year}.csv"
    
    # Load CSVs safely
    try:
        usagers = pd.read_csv(usagers_file, sep=";", encoding="latin1", engine='python', on_bad_lines='skip')
        caracteristiques = pd.read_csv(caract_file, sep=";", encoding="latin1", engine='python', on_bad_lines='skip')
        vehicules = pd.read_csv(veh_file, sep=";", encoding="latin1", engine='python', on_bad_lines='skip')
        lieux = pd.read_csv(lieux_file, sep=";", encoding="latin1", engine='python', on_bad_lines='skip')
    except FileNotFoundError as e:
        print(f"File not found: {e}")
        continue
    
    # Detect merge columns automatically
    num_acc_col = find_column(usagers, ["num_acc", "numacc", "num_accident"])
    num_veh_col = find_column(usagers, ["num_veh", "numveh"])
    
    veh_num_acc_col = find_column(vehicules, ["num_acc", "numacc", "num_accident"])
    veh_num_veh_col = find_column(vehicules, ["num_veh", "numveh"])
    
    lieux_num_acc_col = find_column(lieux, ["num_acc", "numacc", "num_accident"])
    carac_num_acc_col = find_column(caracteristiques, ["num_acc", "numacc", "num_accident"])
    
    # Skip year if any required column is missing
    if not all([num_acc_col, num_veh_col, veh_num_acc_col, veh_num_veh_col, lieux_num_acc_col, carac_num_acc_col]):
        print(f"Skipping year {year}: missing columns in one of the tables")
        continue
    
    # Rename columns consistently
    usagers = usagers.rename(columns={num_acc_col: "num_acc", num_veh_col: "num_veh"})
    vehicules = vehicules.rename(columns={veh_num_acc_col: "num_acc", veh_num_veh_col: "num_veh"})
    lieux = lieux.rename(columns={lieux_num_acc_col: "num_acc"})
    caracteristiques = caracteristiques.rename(columns={carac_num_acc_col: "num_acc"})
    
    # Merge datasets
    try:
        df = usagers.merge(vehicules, on=["num_acc", "num_veh"], how="left")
        df = df.merge(lieux, on="num_acc", how="left")
        df = df.merge(caracteristiques, on="num_acc", how="left")
        df = df.drop_duplicates(subset=["num_acc"])
        print(f"Year {year} merged. Shape: {df.shape}")
    except KeyError as e:
        print(f"Merge error for year {year}: {e}")
        continue
    
    # Append to all_data
    all_data = pd.concat([all_data, df], ignore_index=True)

# Save final merged table
all_data.to_csv(OUTPUT_FILE, index=False)
print(f"\nAll years merged. Final dataset saved at: {OUTPUT_FILE}")
print(f"Final dataset shape: {all_data.shape}")
```
